## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties', [['info', 'governor']])
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
world_bank_projects_json = json.load((open('data/world_bank_projects.json')))
json_df = pd.read_json('data/world_bank_projects.json')
json_df.head()
```

```python
#Check columns
json_df.columns
```

```python
#1. Find the 10 countries with most projects
json_df.countryname.value_counts().head(10)
```

```python
#Africa is a continent. Data of smaller countries in Africa might have been collated.
json_df.countryname.value_counts().head(10).drop("Africa")
```

```python
#2. Find the top 10 major project themes (using column 'mjtheme_namecode')
json_df.mjtheme_namecode
```

```python
#It is in JSON format. Use the json_normalize function.
mjthemes = json_normalize(world_bank_projects_json,"mjtheme_namecode")
```

```python
#Has missing values
mjthemes.name.value_counts().head(10)
```

```python
#3. In 2. above you will notice that some entries have only the 
#   code and the name is missing. Create a dataframe with the missing names filled in.

#Get all rows that aren't blank
mjt = mjthemes[mjthemes.name != ""]
```

```python
#Create a dictionary of codes and names
print(len(mjt.code.unique())) #to check number of unique codes

codedict = {}

for code in mjt.code.unique():
    codedict[code] = mjt[mjt.code == code].name.unique()[0]
codedict
```

```python
#Map code to names and fill in the missing values
mjthemes.name = mjthemes.code.map(lambda x: codedict[x])
```

```python
#We can answer question 2 again now that we have filled in the missing values.
mjthemes.name.value_counts().head(10)
```
